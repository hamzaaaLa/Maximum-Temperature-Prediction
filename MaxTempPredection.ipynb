**BackTest**: cette fonction divise les données météorologiques en périodes de temps, entraîne le modèle sur chaque période et fait des prédictions sur les périodes suivantes, puis retourne un DataFrame contenant les prédictions combinées avec les valeurs réelles pour chaque itération de la validation croisée spécifique aux séries temporelles.

```python
def backtest(weather, model, predictors, start=3650, step=90):
    all_predictions = []

    for i in range(start, weather.shape[0], step):
        train = weather.iloc[:i,:]
        test = weather.iloc[i:(i+step),:]

        model.fit(train[predictors], train["target"])

        preds = model.predict(test[predictors])
        preds = pd.Series(preds, index=test.index)
        #Concatène les valeurs réelles de la variable cible et les prédictions dans un DataFrame appelé combined:
        combined = pd.concat([test["target"], preds], axis=1)
        # Renomme les colonnes du DataFrame combined pour les distinguer clairement entre les valeurs réelles et les prédictions
        combined.columns = ["actual", "prediction"]
        # Calcule la différence absolue entre les prédictions et les valeurs réelles, et stocke ces valeurs dans une nouvelle colonne appelée "diff"
        combined["diff"] = (combined["prediction"] - combined["actual"]).abs()

        all_predictions.append(combined)
    return pd.concat(all_predictions)
```

```python
predictions = backtest(weather, rr, predictors)
```

```python
predictions
```

Cette courbe représente les valeurs de température prédites par le modèle pour les mêmes jours que ceux où les données réelles ont été enregistrées. Idéalement,On consate bien que cette courbe suivre de près la courbe des valeurs réelles, reflétant ainsi la capacité du modèle à anticiper les variations de température

```python
combined=predictions
combined.drop("diff", axis=1, inplace=True)
combined.plot()
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error

mean_absolute_error(predictions["actual"], predictions["prediction"])
```
